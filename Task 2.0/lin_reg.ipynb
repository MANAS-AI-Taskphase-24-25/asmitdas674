## **Getting rid of Extra Column**

```python
print(train_data.isnull().sum())
print("\n**********\n")
print(test_data.isnull().sum())
```

```python
train_data = train_data.dropna()
test_data = test_data.dropna()
```

## **Separating Features and Target Variables**

```python
# Separating features and target in training data
X_train = train_data.drop('Price', axis=1)  # Replacing 'Price' with the actual target column name
y_train = train_data['Price']

# Separating features and target in test data
X_test = test_data.drop('Price', axis=1)
y_test = test_data['Price']
```

## **One Hot Encoding**

```python
print(X_train.dtypes)
```

```python
# One-hot encodeing non-numeric columns
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)

# Aligning the train and testing sets to ensure they have the same columns
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

## **Normalization**

```python
mean = X_train.mean()
std = X_train.std()

X_train = (X_train - mean) / std
X_test = (X_test - mean) / std

X_train
```

```python
# Initialising weights(w) and bias(b)

w = np.zeros(X_train.shape[1])  # One weight for each feature
b = 0  # Bias
```

## **Prediction**

```python
def predict(x, w, b):
    """
    Predict the output of a linear model wx + b.

    Parameters:
    x (numpy array): Input feature vector or matrix.
    w (numpy array): Weights vector.
    b (float): Bias term.

    Returns:
    numpy array: Predicted values.
    """
    return np.dot(x, w) + b
```

## **Cost Computation**

Formula used:



```python
def compute_cost(X, y, w, b):
    """
    Compute the cost function for linear regression.

    Parameters:
    X (numpy array): Input feature data set.
    y (numpy array): Target vector.
    w (numpy array): Weights vector.
    b (float): Bias term.

    Returns:
    float: Cost.
    """
    m = X.shape[0]  # Number of training examples

    # Using np.sum() for efficient cost calculation
    f_wb = np.dot(X, w) + b  # Compute predictions for all examples
    cost = np.sum((f_wb - y) ** 2) / (2 * m)  # Sum squared errors and divide by 2m
    return cost
```

## **Gradient Descent Function**




```python
def eval_grad_desc(X, y, w, b, alpha, num_iters):
    """
    Perform num_iters iterations of gradient descent and update w and b.

    Parameters:
    X (numpy array): Input feature data set.
    y (numpy array): Target vector.
    w (numpy array): Weights vector.
    b (float): Bias term.
    alpha (float): Learning rate.
    num_iters (int): Number of iterations.

    Returns:
    numpy array: Updated weights vector.
    float: Updated bias term.
    list: Cost per iteration.
    """
    m = X.shape[0]  # Number of training examples
    
    cost_history = []  # Track the cost at each iteration

    for i in range(num_iters):
        
        # Compute predictions
        f_wb = np.dot(X, w) + b # Predict for all examples at once

        # Compute gradients
        dj_dw = (1 / m) * np.dot(X.T, (f_wb - y))  # ∂J/∂w ; .dot(x, y): computes dot product of x and y
        dj_db = (1 / m) * np.sum(f_wb - y)         # ∂J/∂b ; syntax of np.sum(): np.sum(arr, axis=None, dtype=None, out=None, keepdims=False)

        # Update parameters using gradient descent
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        # Compute and store the cost
        cost = compute_cost(X, y, w, b)  # Cost J(w, b)
        cost_history.append(cost)

        # Optional: Print cost every 100 iterations
        if i % 100 == 0:
            print(f"Iteration {i}: Cost {cost}")

    return w, b, cost_history
```

## **Training the model**

```python
alpha= 0.01
epoch= 10000     # 1 epoch= 10000 iterations

# Creating corresponding numpy arrays

X_train_np= X_train.to_numpy()
y_train_np= y_train.to_numpy()
X_test_np= X_test.to_numpy()
y_test_np= y_test.to_numpy()

w, b, cost_history= eval_grad_desc(X_train_np, y_train_np, w, b, alpha, epoch)

# Print final weights and bias

print("Trained weights:", w)
print("Trained bias:", b)
```

## **Visualisng the Cost Function**

```python
# Plotting the cost function over iterations

plt.plot(range(len(cost_history)), cost_history)
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost Function Over Iterations")
plt.show()
```

## **Evaluating the Model**



```python
def compute_mse(y_actual, y_pred):
    """
    Compute Mean Squared Error (MSE) between actual and predicted values.

    Parameters:
    y_actual (numpy array): Actual target values.
    y_pred (numpy array): Predicted target values.

    Returns:
    float: Mean Squared Error.
    """

    m = y_actual.shape[0]  # Number of examples

    mse= np.sum((y_actual-y_pred)**2) / m  # Sum of squared errors divided by the number of examples

    return mse
```

```python
# Make predictions on the test set
predictions = np.dot(X_test_np, w) + b

mse= compute_mse(y_test_np, predictions)

print("\nMean Squared Error on Test Data:", mse)
```
