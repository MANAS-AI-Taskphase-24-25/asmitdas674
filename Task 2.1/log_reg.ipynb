## Dividing the Training Dataset into training variables and predictors

## Normalising data

```python
from sklearn.preprocessing import StandardScaler

#Creating a scaler object
scaler = StandardScaler()

#Fitting the scaler object to the training data
X_train_sc= scaler.fit_transform(X_train)
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def fit(data, X, y, iters, alpha):
    
	"""
	'iters': Number of iterations
	'alpha': Learning rate
	"""
	samples, features = X.shape

	weights= np.zeros(features)
	bias= 0

	for i in range(iters):

		linear_pred= np.dot(X, weights) + bias	# f(x)= w*x + b
		act_pred= sigmoid(linear_pred)
		
		dJ_dw= (1/samples) * (np.dot(X.T, (act_pred - y)))
		dJ_db= (1/samples) * np.sum(act_pred - y)

		weights-= alpha * dJ_dw
		bias-= alpha * dJ_db

	return weights, bias
```

```python
def predict(X, weights, bias):

    # Returns either 0 or 1 based on the sigmoid output

    linear_pred= np.dot(X, weights) + bias
    act_pred= sigmoid(linear_pred)
    
    return [1 if pred > 0.5 else 0 for pred in act_pred]
```

## Testing the accuracy of this model

```python
def accuracy(y_pred, y_act):
    
    # Returns the accuracy of the model

    return np.sum(y_pred == y_act) / len(y_act)
```

```python
# Fitting the model

w, b= fit(train_data, X_train_sc, y_train, 10000, 3.0)
train_pred= predict(X_train_sc, w, b)

print("--> Accuracy of the training model: ", accuracy(train_pred, y_train) * 100, "%")
```

```python
# Confusion matrix

from sklearn.metrics import confusion_matrix

cm= confusion_matrix(y_train, train_pred)
print(cm)

# Visualizing the confusion matrix

sb.heatmap(cm, annot= True, fmt= "d", cmap= "Blues")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Cost function
def cost_function(X, y, weights, bias):
    m = X.shape[0]  # number of samples
    linear_pred = np.dot(X, weights) + bias
    predictions = sigmoid(linear_pred)
    cost = - (1 / m) * np.sum(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))
    return cost

# Track the cost at each iteration
costs = []

learning_rate = 3.0
iterations = 10000

w_1 = np.zeros(X_train_sc.shape[1])
b_1= 0

for i in range(iterations):
    # Calculate current cost
    cost = cost_function(X_train_sc, y_train, w_1, b_1)
    costs.append(cost)
    
    # Gradient Descent (update weights and bias)
    linear_pred = np.dot(X_train_sc, w_1) + b_1
    act_pred = sigmoid(linear_pred)
    error = act_pred - y_train
    
    # Gradients
    dw = (1 / X_train_sc.shape[0]) * np.dot(X_train_sc.T, error)
    db = (1 / X_train_sc.shape[0]) * np.sum(error)
    
    # Update weights and bias
    w_1 -= learning_rate * dw
    b_1 -= learning_rate * db

# Plotting the cost function over iterations using Matplotlib
plt.figure(figsize=(10, 6))
plt.plot(range(iterations), costs, color='blue', label='Cost Function')
plt.xlabel('Iterations', fontsize=12)
plt.ylabel('Cost', fontsize=12)
plt.title('Cost Function during Gradient Descent', fontsize=16)
plt.grid(True)
plt.legend()
plt.show()
```
