## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

## Loading & Reading Files

```python
# Load the data

train_data = pd.read_csv('Gotem Pumpkins.csv')
test_data = pd.read_csv('Freyja_Pumpkins.csv')

train_data.head()
```

```python
train_data.drop(["Unnamed: 0"], axis= 1, inplace= True)

train_data.head()
```

## Dividing the Training Dataset into training variables and predictors

```python
X_train= train_data.drop(["Class"], axis= 1)
y_train= train_data["Class"]

print(X_train.head())
print("\n-------------------------------------------\n")
print(y_train.head())
```

## Normalising data

```python
from sklearn.preprocessing import StandardScaler

#Creating a scaler object
scaler = StandardScaler()

#Fitting the scaler object to the training data
X_train_sc= scaler.fit_transform(X_train)
```

```python
# Comparison of the original and scaled data

print(X_train.head())
print("\n-------------------------------------------\n")
print(X_train_sc)
```

## Model for Logistic Regression

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
def fit(data, X, y, iters, alpha):
    
	"""
	'iters': Number of iterations
	'alpha': Learning rate
	"""
	samples, features = X.shape

	weights= np.zeros(features)
	bias= 0

	for i in range(iters):

		linear_pred= np.dot(X, weights) + bias	# f(x)= w*x + b
		act_pred= sigmoid(linear_pred)
		
		dJ_dw= (1/samples) * (np.dot(X.T, (act_pred - y)))
		dJ_db= (1/samples) * np.sum(act_pred - y)

		weights-= alpha * dJ_dw
		bias-= alpha * dJ_db

	return weights, bias
```

```python
def predict(X, weights, bias):

    # Returns either 0 or 1 based on the sigmoid output

    linear_pred= np.dot(X, weights) + bias
    act_pred= sigmoid(linear_pred)
    
    return [1 if pred > 0.5 else 0 for pred in act_pred]
```

## Testing the accuracy of this model

```python
def accuracy(y_pred, y_act):
    
    # Returns the accuracy of the model

    return np.sum(y_pred == y_act) / len(y_act)
```

```python
# Fitting the model

w, b= fit(train_data, X_train_sc, y_train, 10000, 3.0)
train_pred= predict(X_train_sc, w, b)

print("--> Accuracy of the training model: ", accuracy(train_pred, y_train) * 100, "%")
```

```python
# Confusion matrix

from sklearn.metrics import confusion_matrix

cm= confusion_matrix(y_train, train_pred)
print(cm)

# Visualizing the confusion matrix

sb.heatmap(cm, annot= True, fmt= "d", cmap= "Blues")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Confusion Matrix")
plt.show()
```

## Testing the Test Data

```python
test_data = test_data.drop(['Unnamed: 0'], axis=1)
test_data.Class= [1 if type == "Çerçevelik" else 0 for type in test_data.Class]

X_test= test_data.drop(["Class"], axis= 1)
y_test= test_data["Class"]

X_test_sc= scaler.transform(X_test)

test_pred= predict(X_test_sc, w, b)

print("--> Accuracy of the test model: ", accuracy(test_pred, y_test) * 100, "%")
```

```python
# Confusion matrix

cm= confusion_matrix(y_test, test_pred)
print(cm)

# Visualizing the confusion matrix

sb.heatmap(cm, annot= True, fmt= "d", cmap= "Blues")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Confusion Matrix")
plt.show()
```
