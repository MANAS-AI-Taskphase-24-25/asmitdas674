## Training Model

```python
# Finding closest centroids using KMeans method

def find_closest_centroids(X, centroids):
    
    k= centroids.shape[0]
    idx = np.zeros(X.shape[0], dtype=int)   # idx is a vector of indices corresponding to the closest centroid for each data point

    for i in range(X.shape[0]):  # Iterating over all data points in X (i.e., all rows in X)
        
        distance= []    # Array to hold distance between X[i] and each centroids[j] for every iteration
                        # After the previous iteration is done, distance gets reinitialised for the next iteration

        for j in range(k):  # Iterating over all centroids (i.e., all rows in centroids)
            
            norm_ij = np.linalg.norm(X[i] - centroids[j])   

            """
                linaglg: A NumPy module to carry out linear arithmetic operations like norm here
                         (i.e., norm= sqrt(x1[i]^2 - x2[2]^2))
                                                                
            """
            distance.append(norm_ij)

        idx[i] = np.argmin(distance)   # Assigning the index of the centroid closest to X[i] to idx[i]

    return idx

def compute_centroids(X, idx, k):
    
    n= X.shape[1]   # n: Number of features in the dataset (basically, number of columns in X)
    centroids= np.zeros((k, n))   # Array to hold the new centroids
    
    for i in range(k):
        
        centroids[i] = np.mean(X[idx == i], axis=0)   # Computing the mean of all data points assigned to the ith centroid

    return centroids
```

## Fitting our data to the model

```python
iterations = 10
k = 3

# Randomly initializing centroids
centroids = np.random.rand(k, data_pca.shape[1]) # centroids is a 3x10 matrix (since data_pca has 10 features)

for i in range(iterations):

    idx = find_closest_centroids(data_pca, centroids)
    centroids = compute_centroids(data_pca, idx, k)

print(centroids)
```

```python
# Visualizing the clusters

fig = plt_.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_pca[idx == 0, 0], data_pca[idx == 0, 1], data_pca[idx == 0, 2], c='blue', marker='o', label='Cluster 1')
ax.scatter(data_pca[idx == 1, 0], data_pca[idx == 1, 1], data_pca[idx == 1, 2], c='red', marker='o', label='Cluster 2')
ax.scatter(data_pca[idx == 2, 0], data_pca[idx == 2, 1], data_pca[idx == 2, 2], c='green', marker='o', label='Cluster 3')
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], c='black', marker='x', s=200, label='Centroids')
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt_.title("KMeans Clustering")
plt_.legend()
plt_.show()
```

## Evaluating our model

```python
def compute_inertia(X, idx, centroids):
    inertia = 0
    for i in range(len(centroids)):
        cluster_points = X[idx == i]
        inertia += np.sum(np.linalg.norm(cluster_points - centroids[i], axis=1) ** 2)
    return inertia

inertia= compute_inertia(data_pca, idx, centroids)
print(inertia)
```

```python
def run_kmeans(X, initial_centroids, max_iters=100):
    """
    Runs the KMeans algorithm on data X.

    Parameters:
    - X: Data points (array of shape (m, n)).
    - initial_centroids: Initial centroids (array of shape (k, n)).
    - max_iters: Maximum number of iterations to run KMeans.

    Returns:
    - centroids: Final centroids after convergence.
    - idx: Cluster assignments for each data point.
    """
    centroids = initial_centroids
    for _ in range(max_iters):
        # Step 1: Assign data points to the closest centroid
        idx = find_closest_centroids(X, centroids)
        
        # Step 2: Recompute centroids
        new_centroids = compute_centroids(X, idx, len(initial_centroids))
        
        # Check for convergence (if centroids don't change, stop)
        if np.all(centroids == new_centroids):
            break
        
        centroids = new_centroids  # Update centroids
    
    return centroids, idx


# Function to initialize centroids randomly
def initialize_centroids(X, k):
    indices = np.random.choice(X.shape[0], size=k, replace=False)
    return X[indices]

# Run KMeans and compute inertia for different values of k
k_values = range(1, 11)  # k from 1 to 10
inertias = []

for k in k_values:
    initial_centroids = initialize_centroids(data_pca, k)
    centroids, idx = run_kmeans(data_pca, initial_centroids, 10)
    inertia = compute_inertia(data_pca, idx, centroids)
    inertias.append(inertia)

# Plot inertia vs k
plt_.plot(k_values, inertias, marker='o')
plt_.xlabel('Number of Clusters (k)')
plt_.ylabel('Inertia')
plt_.title('Elbow Method for Optimal k')
plt_.show()
```

Looking at the graph above, we can conclude that our choice of keeping k= 3 was correct as the elbow bend occurs at k= 3 here.

```python
import sklearn.cluster as sk

# Initialize an empty dataframe
df = pd.DataFrame(data_pca)

for k in range(1, 11):
    kmeans = sk.KMeans(n_clusters=k, random_state=0).fit(data_pca)
    
    # Add the cluster labels to the dataframe as a new column
    df[f"KMeans_{k}"] = kmeans.labels_

df
```

```python
fig, axes = plt_.subplots(5, 2, figsize=(20, 20))

for i, ax in enumerate(axes.flatten()):

    ax.scatter(df[0], df[1], c=df[f"KMeans_{i+1}"], cmap='viridis', marker='o') # cmap is the colour map used to colour the clusters
    ax.set_title(f"KMeans Clustering with k={i+1}")
    ax.set_xlabel("PC1")
    ax.set_ylabel("PC2")

plt_.tight_layout()
plt_.show()
```
